```python
import matplotlib.pyplot as plt

def calculate_real_signatures():
    # Valós aláírási idők másodpercben
    rsa_real_time = 0.003  # RSA aláírás ideje (3 ms)
    dsa_real_time = 0.002  # DSA aláírás ideje (2 ms)
    ecdsa_real_time = 0.0008  # ECDSA aláírás ideje (0.8 ms)
    sphincs_real_time = 0.07  # SPHINCS+ aláírás ideje (70 ms)
    xmss_real_time = 0.06  # XMSS aláírás ideje (60 ms)
    crystals_dilithium_real_time = 0.0025  # CRYSTALS-Dilithium aláírás ideje (2.5 ms)
    falcon_real_time = 0.0015  # Falcon aláírás ideje (1.5 ms)

    # Aláírások másodpercenként
    rsa_signatures = 1 / rsa_real_time
    dsa_signatures = 1 / dsa_real_time
    ecdsa_signatures = 1 / ecdsa_real_time
    sphincs_signatures = 1 / sphincs_real_time
    xmss_signatures = 1 / xmss_real_time
    crystals_dilithium_signatures = 1 / crystals_dilithium_real_time
    falcon_signatures = 1 / falcon_real_time

    # Kiírás
    print("\n--- Valós aláírások másodpercenként ---")
    print(f"RSA: {rsa_signatures:.2f} aláírás/másodperc ({rsa_real_time:.4f} másodperc/aláírás)")
    print(f"DSA: {dsa_signatures:.2f} aláírás/másodperc ({dsa_real_time:.4f} másodperc/aláírás)")
    print(f"ECDSA: {ecdsa_signatures:.2f} aláírás/másodperc ({ecdsa_real_time:.4f} másodperc/aláírás)")
    print(f"SPHINCS+: {sphincs_signatures:.2f} aláírás/másodperc ({sphincs_real_time:.4f} másodperc/aláírás)")
    print(f"XMSS: {xmss_signatures:.2f} aláírás/másodperc ({xmss_real_time:.4f} másodperc/aláírás)")
    print(f"CRYSTALS-Dilithium: {crystals_dilithium_signatures:.2f} aláírás/másodperc ({crystals_dilithium_real_time:.4f} másodperc/aláírás)")
    print(f"Falcon: {falcon_signatures:.2f} aláírás/másodperc ({falcon_real_time:.4f} másodperc/aláírás)")

    # Eredmények az oszlopdiagramhoz
    algorithms = ["RSA", "DSA", "ECDSA", "SPHINCS+", "XMSS", "CRYSTALS-Dilithium", "Falcon"]
    signatures_per_second = [
        rsa_signatures, dsa_signatures, ecdsa_signatures, sphincs_signatures,
        xmss_signatures, crystals_dilithium_signatures, falcon_signatures
    ]

    # Oszlopdiagram készítése
    plt.figure(figsize=(14, 6))
    plt.bar(algorithms, signatures_per_second, color=['blue', 'orange', 'green', 'purple', 'pink', 'red', 'yellow'])
    plt.title("Digitális aláírások teljesítményének összehasonlítása")
    plt.ylabel("Aláírások száma (1 másodperc alatt)")
    plt.xlabel("Algoritmusok")
    plt.grid(axis="y", linestyle="--", linewidth=0.7)
    plt.xticks(rotation=15)
    plt.show()

if __name__ == "__main__":
    calculate_real_signatures()
```
